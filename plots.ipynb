# Notebook for post plots

```python
MLENERGY_GREEN = "#23d175"
MLENERGY_BG = "#2e303e"
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from IPython.display import display

def savesvg(fig_fn, prefix, **kwargs):
    """Save the figure to SVG in a git-friendly manner, in both dark and light colors."""
    light_svg_path = f"docs/posts/assets/{prefix}-light.svg"
    dark_svg_path = f"docs/posts/assets/{prefix}-dark.svg"
    
    with plt.style.context("default"):
        fig = fig_fn()
        display(fig)
        fig.savefig(light_svg_path, metadata={"Date": None}, transparent=True, **kwargs)
        
    with plt.style.context("dark_background"):
        fig = fig_fn()
        display(fig)
        fig.savefig(dark_svg_path, metadata={"Date": None}, transparent=True, **kwargs)
    
    print("Saved to")
    print(light_svg_path)
    print(dark_svg_path)
```

## Measuring GPU energy consumption: Best practices
https://ml.energy/blog/energy/measurement/measuring-gpu-energy-best-practices/

```python
def draw():
    # Data (from zeus/trace/summary_power_v100.csv)
    models = ['ShuffleNetV2', 'ResNet-50', 'DeepSpeech2', 'NeuMF', 'Bert Base']
    power_consumption = [171.261019, 217.536769, 160.558409, 36.218861, 226.331563]

    fig, ax = plt.subplots(tight_layout=True, figsize=(6, 3))
    ax.bar(models, power_consumption, color=MLENERGY_GREEN)
    ax.axhline(250, color=MLENERGY_GREEN, linestyle='--')
    ax.text(4.75, 242, 'TDP', color=MLENERGY_GREEN)
    ax.set_ylabel('Average Power (W)')
    ax.set_title('Average Power Consumption on V100')
    
    ax.set_ylim(0, 320)
    
    return fig

savesvg(draw, "correctly-measuring-energy/avg-power-v100")
```
